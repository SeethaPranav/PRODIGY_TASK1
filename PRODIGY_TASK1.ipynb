# Bar Plot

```python
season_map = {1: 'Spring', 2: 'Summer', 3: 'Fall', 4: 'Winter'}
data['season_label'] = data['season'].map(season_map)

sns.countplot(x='season_label', data=data, palette='coolwarm')
plt.title('Distribution of Seasons')
plt.xlabel('Season')
plt.ylabel('Count')
plt.show()
```

# Histogram

```python
plt.figure(figsize=(8, 6))
sns.histplot(data['cnt'], bins=30, kde=True, color='teal')  
plt.title('Distribution of Bike Rental Counts')
plt.xlabel('Count')
plt.ylabel('Frequency')
plt.show()
```
